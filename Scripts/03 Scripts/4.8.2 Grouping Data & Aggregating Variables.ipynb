### 04. Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Split the data into groups based on the “user_id” column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(20)
```

### Deriving Columns with loc()

```python
# Create flags to assign "loyalty" labels to user ID based on max order value
ords_prods_merge.loc[ords_prods_merge['max_order']> 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order']<= 40) & (ords_prods_merge['max_order']> 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Check frequencies of loyalty flags for accuracy
ords_prods_merge['loyalty_flag'].value_counts()
```

### 05. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category

### 06. Create a spending flag for each user based on the average price across all their orders

```python
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Create flags to assign "spending" labels
ords_prods_merge.loc[ords_prods_merge['avg_price']>= 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price']< 10, 'spending_flag'] = 'Low Spender'
```

```python
# Check frequencies
ords_prods_merge['spending_flag'].value_counts()
```

### 07. Create and order frequency flag that marks the regularity of a user's ordering behaviour according to the median in the "days_since_prior_order" column

```python
#Gorup data by user_id and create a 'media_days' column
ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Check grouping
ords_prods_merge.head()
```

```python
# Create flags to assign "freqency" labels 
ords_prods_merge.loc[ords_prods_merge['median_days']> 20, 'frequency_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days']<= 20) & (ords_prods_merge['median_days']> 10), 'frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days']<= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['frequency_flag'].value_counts()
```
